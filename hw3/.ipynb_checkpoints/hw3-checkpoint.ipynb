```python
def sigmoid(z):   
    return 1 / (1 + np.exp(-z))  

# def log_likelihood(fea, tar, wei):
#     sco = np.dot(fea, wei)
#     l = np.sum(tar * sco - np.log(1 + np.exp(sco)))
#     return l

def logLiklihood(z, y):
    return -1 * np.sum((y * np.log(sigmoid(z))) + ((1 - y) * np.log(1 - sigmoid(z))))

def reg_logLiklihood(x, weights, y, C):
    z = np.dot(x, weights) 
    reg_term = (1 / (2 * C)) * np.dot(weights.T, weights)
    return -1 * np.sum((y * np.log(logistic_func(z))) + ((1 - y) * np.log(1 - logistic_func(z)))) + reg_term

class myknn_regressor:
    def __init__(self, reg_vec, max_iter, tol, add_intercept = False):
        self.reg_vec = np.diag(reg_vec)
        self.max_iter = max_iter
        self.tol = tol
        self.add_intercept = add_intercept
        
    def fit(self, X_tr, Y_tr):
        self.X_tr = X_tr
        self.Y_tr = Y_tr
        
        if self.add_intercept == True:
#         (1, f+1)
            X_train = np.c_[self.X_tr, np.ones([np.shape(self.X_tr)[0], 1])]
#         (n, f+1)
        else:
            X_train = np.array(self.X_tr)
#         print(np.dot(X_train.T, X_train).shape)
#         print(np.mean(self.reg_vec)*np.diag(np.array([1]*np.shape(X_train)[1])))
        self.weights = np.dot(np.dot(np.linalg.inv(
            np.dot(X_train.T, X_train) + np.mean(self.reg_vec)*np.diag(np.array([1]*np.shape(X_train)[1]))
        ), X_train.T), self.Y_tr)
#         print(self.weights)
        for i in range(self.max_iter):
#             print("iter:")
#             print(i)
#             print(self.weights)
            
#             grad = self.reg_vec * self.weights
            z = np.dot(X_train, self.weights)
            grad = np.dot(self.reg_vec, self.weights) + np.dot(X_train.T, (sigmoid(z) - self.Y_tr))
    
#             print(grad.shape)
#             errors = self.Y_tr - sigmoid(z)
#             grad += np.dot(errors, X_train)
            R = sigmoid(z)*(1-sigmoid(z))
#             Rnn = np.diag(R)
            Hess = np.dot(X_train.T*R, X_train) + self.reg_vec
#             print(grad.shape)
#             print(np.linalg.inv(Hess).shape)
#             del R
            delta_w = np.dot(np.linalg.inv(Hess), grad)
            if np.all(abs(delta_w) >= self.tol): 
                self.weights -= delta_w  
            else:
                break
                
        return self
        
    def predict(self, X_t, p = True):
        np.set_printoptions(threshold=np.inf)
        if p == True:
            print(self.weights)
        else:
            pass
        z = self.weights[len(self.weights)-1] + np.dot(X_t, self.weights[:-1]) 
#         print(z)
        probs = np.array([sigmoid(i) for i in z])
        predictions = np.where(probs >= 0.5, 1, 0)
        return predictions, probs
    
    
```
