```python
import pickle
import math
import numpy as np
from sklearn import preprocessing
with open('msd_data1.pickle', 'rb') as f:
     data = pickle.load(f)
# print(data)
X_train = data['X_train']
X_test = data['X_test']
Y_train = data['Y_train']
Y_test = data['Y_test']
X_tra = preprocessing.scale(X_train)
X_tes = preprocessing.scale(X_test)
```
