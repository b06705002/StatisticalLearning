```python
import pickle
import numpy as np
```

```python
with open('msd_data1.pickle','rb') as x:
    data = pickle.load(x)
print(data.keys())
X_train = data['X_train']
Y_train = data['Y_train']
X_test = data['X_test']
Y_test = data['Y_test']
stdXTR = (X_train-X_train.mean(axis=0))/X_train.std(axis=0)
stdXTE = (X_test-X_test.mean(axis=0))/X_test.std(axis=0)
print(type(X_test))
#print(type(data['X_train']))
#print(data['Y_test'][0:20])
```

```python
class myknn_regressor:
    
    def __init__(se,kk,method):
        se.k = kk
        se.method = method
    def fit(se,XT,YT):
        se.XT = XT
        se.YT = YT
    def predict(se,XTEST):
        #print(se.stdX)
        se.XTEST = XTEST
        y_pred = []
        for i in range(len(se.XTEST)):
            dist = []
            dist = np.sum(((se.XT - se.XTEST[i]) ** 2),axis=1)
            
            NNY = []
            for j in range(20):
                p = np.argmin(dist)
                NNY.append(se.YT[p])
                dist[p] = float('INF')
            npNNY = np.array(NNY)
            
            print(npNNY)
            newNNY = []
            out = 0
            if se.method == 'remove_outliers':
                out = 1
                q1 = np.quantile(npNNY, .25)
                q3 = np.quantile(npNNY, .75)
                IQR = q3-q1
                for l in range(20):
                    if (npNNY[l] >= (q1-IQR*1.5)) and (npNNY[l] <= (q3+IQR*1.5)):
                        newNNY.append(npNNY[l])
            if out == 0:
                y_pred.append(sum(NNY)/len(NNY))
            else:
                y_pred.append(sum(newNNY)/len(newNNY))
            npy = np.array(y_pred)
            npy = np.around(npy)
        print('predict y =',npy[0:20])
        RMSE = (((npy - Y_test) ** 2).sum() / len(npy)) ** (1/2)
        print('RMSE is ',RMSE)
    

myknn = myknn_regressor(20,'equal_weight')
myknn.fit(stdXTR,Y_train)
myknn.predict(stdXTE)
'''
predict y = [1993. 1994. 2000. 1992. 1993. 1998. 1987. 1991. 2002. 2003. 2001. 1999.
 1996. 1997. 1995. 1997. 1992. 1999. 2004. 1996.]
RMSE is  10.29443215205838
'''
```

```python
from sklearn.neighbors import KNeighborsRegressor
import matplotlib.pyplot as plt
k = [1,2,3,4,5,10,15,20,25,30,35,40,45,50,55,60,80,100,120,140,160,180,200]
k = np.array(k)
stdXTR = (X_train-X_train.mean(axis=0))/X_train.std(axis=0)
stdXTE = (X_test-X_test.mean(axis=0))/X_test.std(axis=0)
RMSE_std = []
RMSE_not_std = []
RMSE_myknn = []

for i in k:
    knn = KNeighborsRegressor(n_neighbors=i,algorithm='brute')
    knn.fit(stdXTR,Y_train)
    predict_y_std = knn.predict(stdXTE)
    predict_y_std = np.around(predict_y_std)
    RMSE_tmp = (((predict_y_std - Y_test) ** 2).sum() / len(Y_test)) ** (1/2)
    RMSE_std.append(RMSE_tmp)

RMSE_std = np.array(RMSE_std)

for i in k:
    knn = KNeighborsRegressor(n_neighbors=i,algorithm='brute')
    knn.fit(X_train,Y_train)
    predict_y_std = knn.predict(X_test)
    predict_y_std = np.around(predict_y_std)
    RMSE_tmp = (((predict_y_std - Y_test) ** 2).sum() / len(Y_test)) ** (1/2)
    RMSE_not_std.append(RMSE_tmp)

RMSE_not_std = np.array(RMSE_not_std)

class myknn_regressor:
    
    def __init__(se,kk,method):
        se.k = kk
        if kk < 10:
            se.method = 'equal_weight'
        else:
            se.method = method
    def fit(se,XT,YT):
        se.XT = XT
        se.YT = YT
    def predict(se,XTEST):
        #print(se.stdX)
        se.XTEST=XTEST
        y_pred = []
        for i in range(len(se.XTEST)):
            dist = []
            dist = np.sum(((se.XT - se.XTEST[i]) ** 2),axis=1)
            
            NNY = []
            for j in range(20):
                p = np.argmin(dist)
                NNY.append(se.YT[p])
                dist[p] = float('INF')
            npNNY = np.array(NNY)
            newNNY = []
            out = 0
            if se.method == 'remove_outliers':
                out = 1
                q1 = np.quantile(npNNY, .25)
                q3 = np.quantile(npNNY, .75)
                IQR = q3-q1
                for l in range(20):
                    if (npNNY[l] >= (q1-IQR*1.5)) and (npNNY[l] <= (q3+IQR*1.5)):
                        newNNY.append(npNNY[l])
            if out == 0:
                y_pred.append(sum(NNY)/len(NNY))
            else:
                y_pred.append(sum(newNNY)/len(newNNY))
            npy = np.array(y_pred)
        return npy


for i in k:
    myknn = myknn_regressor(i,'remove_outliers')
    myknn.fit(stdXTR,Y_train) 
    predict_y_std = myknn.predict(stdXTE)
    predict_y_std = np.around(predict_y_std)
    RMSE_tmp = (((predict_y_std - Y_test) ** 2).sum() / len(Y_test)) ** (1/2)
    RMSE_myknn.append(RMSE_tmp)

RMSE_myknn = np.array(RMSE_myknn)

case_one_plot = plt.scatter(k,RMSE_std)
case_two_plot = plt.scatter(k,RMSE_not_std)
case_three_plot = plt.scatter(k,RMSE_myknn)
plt.legend(handles=[case_one_plot,case_two_plot,case_three_plot],labels=['case_one_RMSE','case_two_RMSE','case_three_RMSE'])
plt.show()


```
